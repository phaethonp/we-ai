```python
import json
import pandas as pd
```

```python
df = pd.read_json('/content/sample_data/Duplex_A_20110907_optimized.json')
df
```

```python
data = pd.json_normalize(df['data'], meta=df.columns.drop('data').tolist())
data
```

```python
# List to store the names of columns containing lists
list_columns = []

# Iterate over the columns of the DataFrame
for col in data.columns:
    # Check if at least two elements in the column are lists
    if sum(isinstance(element, list) for element in data[col]) >= 2:
        list_columns.append(col)

list_columns
```

```python
# Create a new DataFrame to store the flattened columns
new_data = {}

# Flatten the columns containing lists
for col in list_columns:
    # Check if the column contains lists
    if all(isinstance(element, list) for element in data[col]):
        # Flatten the column by concatenating the elements
        flattened_values = [item for sublist in data[col] for item in sublist]

        # Store the flattened values in the new DataFrame
        new_data[col] = flattened_values

# Add non-list columns from the original DataFrame to the flattened DataFrame
for col in data.columns:
    if col not in list_columns:
        new_data[col] = data[col]

# Create the flattened DataFrame
new_data = pd.DataFrame(new_data)
new_data
```

```python
nan_count = new_data.isna().sum()
nan_count
```

```python
import pandas as pd

# Threshold for NaN percentage to drop columns
threshold = 0.8  # 80% NaN
# Calculate the NaN percentage in each column
nan_percentage = new_data.isna().mean()

# List of columns to drop
columns_to_drop = nan_percentage[nan_percentage > threshold].index.tolist()

# Drop the columns from the DataFrame
new_data1 = new_data.drop(columns=columns_to_drop)
new_data1
```

```python
# Delete a column from the original DataFrame
df1 = df.drop('data', axis=1)

# Concatenate the two DataFrames
df_final = pd.concat([df1, new_data1], axis=1)
df_final
```

```python
# Save the DataFrame to a CSV file
df_final.to_csv('data.csv', index=False)
```
